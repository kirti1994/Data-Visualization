```python
import pandas as pd
```

```python
df=pd.read_csv('athlete_events.csv')
```

```python
df.head()
```

```python
df_1=df[['Year', 'Country', 'Sport', 'Medal', 'NOC']]
```

```python
df_1.head()
```

```python
df_noc=pd.read_csv('noc_regions.csv')
```

```python
df_noc.head()
```

```python
df_1_noc = pd.merge(df_1, df_noc, how='left', left_on = 'NOC', right_on = 'NOC')
```

```python
df_1_noc.head()
```

```python
df_1_noc = df_1_noc.drop(['NOC', 'notes'],axis = 1)
```

```python
df_1_noc=df_1_noc.rename(columns={"region": "PlayerCountry"})
```

```python
# Get one hot encoding of columns B
one_hot = pd.get_dummies(df_1_noc['Medal'])
# Drop column B as it is now encoded
df_1_noc = df_1_noc.drop('Medal',axis = 1)
# Join the encoded df
df_1_noc = df_1_noc.join(one_hot)
df_1_noc.head()  
```

```python
grouped_multiple = df_1_noc.groupby(['Year', 'Country', 'Sport', 'PlayerCountry']).agg({'Bronze': ['sum'], 'Gold': ['sum'], 'Silver': ['sum']})
grouped_multiple.columns = ['Bronze', 'Gold', 'Silver']
grouped_multiple = grouped_multiple.reset_index()
print(grouped_multiple)
```

```python
df_1_noc=pd.DataFrame(grouped_multiple)
```

```python
df_1_noc = df_1_noc.drop(['Country'],axis = 1)
```

```python
df_1_noc.head()
```

```python
df_1_noc.to_csv('tableau.csv', index=False)
```
